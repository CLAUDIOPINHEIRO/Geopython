````python
def make_multipoint(df, reproject=False):
    """
    Return a Shapely Multipoint from a DataFrame
    Assumes that longitude and latitude columns exist
    """
    # we don't want any devices missing data
    # TODO: we need better data validation
    wgs84 = pyproj.Proj("+init=EPSG:4326")
    osgb36 = pyproj.Proj("+init=EPSG:27700")
    df[['lon', 'lat']] = df[['lon', 'lat']].astype(float)
    tpl = zip(df['lon'], df['lat'])
    if reproject:
        x, y = pyproj.transform(wgs84, osgb36, list(df['lon']), list(df['lat']))
        tpl = zip(x, y)
    return MultiPoint(tpl)

def dump_geojson(df, fname):
    """
    Dump Pandas DataFrame long and lats to a GeoJSON file
    """
    mp = make_multipoint(df)
    with open('data/' + fname, 'w') as out:
        out.write(geojson.dumps(mp))
        out.write('\n')

def available_properties(df, city):
    """ Return available sensor properties for a city """
    l = list()
    for each in df[df.city == city]['properties'].iteritems():
        for elem in each[1]:
            l.append(elem)
    return list(set(l))

def colorbar_index(ncolors, cmap, labels=None, **kwargs):
    """
    This is a convenience function to stop you making off-by-one errors
    Takes a standard colourmap, and discretizes it,
    then draws a color bar with correctly aligned labels
    """
    cmap = cmap_discretize(cmap, ncolors)
    mappable = cm.ScalarMappable(cmap=cmap)
    mappable.set_array([])
    mappable.set_clim(-0.5, ncolors+0.5)
    colorbar = plt.colorbar(mappable, **kwargs)
    colorbar.set_ticks(np.linspace(0, ncolors, ncolors))
    colorbar.set_ticklabels(range(ncolors))
    if labels:
        colorbar.set_ticklabels(labels)
    return colorbar

def cmap_discretize(cmap, N):
    """
    Return a discrete colormap from the continuous colormap cmap.
    
        cmap: colormap instance, eg. cm.jet. 
        N: number of colors.
    
    Example
        x = resize(arange(100), (5,100))
        djet = cmap_discretize(cm.jet, 5)
        imshow(x, cmap=djet)

    """
    if type(cmap) == str:
        cmap = get_cmap(cmap)
    colors_i = concatenate((linspace(0, 1., N), (0., 0., 0., 0.)))
    colors_rgba = cmap(colors_i)
    indices = linspace(0, 1., N + 1)
    cdict = {}
    for ki,key in enumerate(('red','green','blue')):
        cdict[key] = [(indices[i], colors_rgba[i - 1, ki], colors_rgba[i, ki]) for i in xrange(N + 1)]
    return matplotlib.colors.LinearSegmentedColormap(cmap.name + "_%d" % N, cdict, 1024)
````

**If we want to show quantile *intervals***

`bins` includes the *lowest quantiled value*, so it's `categorical.levels` + 1 in length, whereas `categorical.levels` returns intervals

Since we probably don't care about the lowest value, we can remove it from our labels.

If we wanted to label our colour bar with intervals, we'd use `categorical.levels`
```python
categorical, bins = pd.qcut(df_map['density'], 7, retbins=True)
display = []
for k in categorical.levels:
    ivl = k.split(',')
    display.append([float(ivl[0][1:]), float(ivl[1][:-1])])
display = sorted(["%0.8f - %0.8f/$ft^2$" % (hi, lo) for hi, lo in display])
display.insert(0, "No Sensors")
labels = display
```

````python
# Calculate quantiles for density
categorical, bins = pd.qcut(df_map['density_km'], 5, retbins=True)
df_map['quantile_bins'] = categorical.labels
````

````python
# Calculate Jenks natural breaks for density
breaks = nb(
    df_map[df_map['density_km'].notnull()].density_km.values,
    initial=300,
    k=5)
jb = pd.DataFrame({'jenks_bins': breaks.yb}, index=df_map[df_map['density_km'].notnull()].index)
df_map = df_map.join(jb)
df_map.jenks_bins.fillna(-1, inplace=True)
````

````python
quantile_labels = ["<= %0.2f/$km^2$(%s wards)" % (b, w) for b, w in zip(
    bins[1:],
    [k for k in pd.value_counts(categorical, sort=False)])]
quantile_labels.insert(0, 'No sensors (%s wards)' % len(df_map[df_map['density_km'].isnull()]))
jenks_labels = ["<= %0.1f/km$^2$(%s wards)" % (b, w) for b, w in zip(
    breaks.bins, breaks.counts)]
jenks_labels.insert(0, 'No plaques (%s wards)' % len(df_map[df_map['density_km'].isnull()]))
````
